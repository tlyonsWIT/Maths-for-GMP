```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import cos, sin, tan, exp, pi, sqrt, log
```

```python
from matplotlib import rc
rc('text',usetex=True)
```

```python
mu1=0
mu2=1.5
mu3=-0.4
```

```python
sd1=1
sd2=0.5
sd3=2
```

```python
def f(x,mu,sd):
    return 1/(sqrt(2*pi*sd))*exp(-0.5*(x-mu)**2/sd**2)
```

```python
x=np.linspace(mu1-2*sd3,mu1+2*sd3,1001)
```

```python
plt.plot(x,f(x,mu1,sd1),color='red',label='$\mu=0$ $\sigma=1$')
plt.plot(x,f(x,mu1,sd2),color='blue',label='$\mu=0$ $\sigma=0.5$')
plt.plot(x,f(x,mu1,sd3),color='darkslategray',label='$\mu=0$ $\sigma=2$')
plt.legend()
```

```python
x=np.linspace(mu3-2*sd3,mu2+2*sd3,1001)
plt.plot(x,f(x,mu1,sd1),color='red',label='$\mu=0$ $\sigma=1$')
plt.plot(x,f(x,mu2,sd2),color='blue',label='$\mu=1.5$ $\sigma=0.5$')
plt.plot(x,f(x,mu3,sd3),color='darkslategray',label='$\mu=-0.4$ $\sigma=2$')
plt.legend()
```
